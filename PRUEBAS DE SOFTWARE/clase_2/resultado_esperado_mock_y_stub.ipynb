## Ejercicio 2

**Solución:** usar un mock como stub configurando `return_value` en `obtener_precio`.

**Resultado esperado:** `calcular_total(..., 4)` devuelve `322.0`.

```python
from unittest.mock import Mock

def calcular_total(repositorio, cantidad):
    precio = repositorio.obtener_precio("SKU-1")
    return precio * cantidad

repositorio = Mock()
repositorio.obtener_precio.return_value = 80.5

total = calcular_total(repositorio, 4)
assert total == 322.0
print(f"Ejercicio 2 completado. Total = {total}")
```

## Ejercicio 3

**Solución:** reemplazar `requests.get` con `patch`, simular un `json()` de respuesta y validar el valor retornado.

**Resultado esperado:** la función devuelve `Lucia` y se imprime `Ejercicio 3 completado`.

```python
from unittest.mock import patch
import requests

def obtener_nombre():
    respuesta = requests.get("https://api.ejemplo.com/usuario")
    return respuesta.json()["nombre"]

with patch("requests.get") as mock_get:
    mock_get.return_value.json.return_value = {"nombre": "Lucia"}
    nombre = obtener_nombre()
    assert nombre == "Lucia"
    mock_get.assert_called_once_with("https://api.ejemplo.com/usuario")

print("Ejercicio 3 completado")
```

## Ejercicio 4

**Solución:** configurar `side_effect` para que `guardar` lance una excepción controlada.

**Resultado esperado:** la función no propaga el error y devuelve `{"estado": "error", "detalle": "fallo_guardado"}`.

```python
from unittest.mock import Mock

class ErrorGuardado(Exception):
    pass

def registrar(repositorio, pedido):
    try:
        repositorio.guardar(pedido)
        return {"estado": "ok"}
    except ErrorGuardado:
        return {"estado": "error", "detalle": "fallo_guardado"}

repositorio = Mock()
repositorio.guardar.side_effect = ErrorGuardado("fallo al guardar")

resultado = registrar(repositorio, {"id": 10})
assert resultado == {"estado": "error", "detalle": "fallo_guardado"}
print(resultado)
```

## Ejercicio 5

**Solución:** usar `MagicMock` para simular un objeto compatible con `with` y configurar `leer()`.

**Resultado esperado:** la función devuelve `contenido de prueba`.

```python
from unittest.mock import MagicMock

def cargar_texto(recurso):
    with recurso as archivo:
        return archivo.leer()

recurso = MagicMock()
recurso.__enter__.return_value = recurso
recurso.leer.return_value = "contenido de prueba"

texto = cargar_texto(recurso)
assert texto == "contenido de prueba"
print(texto)
```

## Ejercicio 6

**Solución:** aplicar `@patch("requests.get")`, simular el `json` del mock y ejecutar el test con `unittest`.

**Resultado esperado:** el test `test_temperatura_bogota` pasa correctamente.

```python
import unittest
from unittest.mock import patch

def obtener_temperatura(ciudad):
    import requests
    respuesta = requests.get(f"https://api.clima.com/?ciudad={ciudad}")
    return respuesta.json()["temperatura"]

class TestClimaPractica(unittest.TestCase):

    @patch("requests.get")
    def test_temperatura_bogota(self, mock_get):
        mock_get.return_value.json.return_value = {"temperatura": 18}

        temperatura = obtener_temperatura("Bogota")

        self.assertEqual(temperatura, 18)
        mock_get.assert_called_once_with("https://api.clima.com/?ciudad=Bogota")

suite = unittest.TestLoader().loadTestsFromTestCase(TestClimaPractica)
runner = unittest.TextTestRunner(verbosity=2)
runner.run(suite)
```

## Ejercicio 7

**Solución:** crear un mock del servicio, ejecutar tres llamadas y validar `call_count` y `call_args_list`.

**Resultado esperado:** se registran exactamente estas llamadas: `("A", 100)`, `("B", 200)` y `("C", 300)`.

```python
from unittest.mock import Mock

def notificar_pagos(pagos, servicio):
    for pago in pagos:
        servicio.enviar(pago["cliente"], pago["valor"] )

pagos = [
    {"cliente": "A", "valor": 100},
    {"cliente": "B", "valor": 200},
    {"cliente": "C", "valor": 300},
]

servicio = Mock()
notificar_pagos(pagos, servicio)

assert servicio.enviar.call_count == 3
llamadas = servicio.enviar.call_args_list
assert llamadas[0].args == ("A", 100)
assert llamadas[1].args == ("B", 200)
assert llamadas[2].args == ("C", 300)

print(llamadas)
```

## Ejercicio 8

**Solución:** usar `patch.object` para reemplazar `obtener_impuesto` en una clase real sin modificar su implementación original.

**Resultado esperado:** `total(300)` devuelve `357`.

```python
from unittest.mock import patch

class CalculadoraImpuestos:
    def obtener_impuesto(self, base):
        raise NotImplementedError("Depende de un servicio externo")

    def total(self, base):
        return base + self.obtener_impuesto(base)

calculadora = CalculadoraImpuestos()

with patch.object(CalculadoraImpuestos, "obtener_impuesto", return_value=57):
    total = calculadora.total(300)

assert total == 357
print(total)
```
