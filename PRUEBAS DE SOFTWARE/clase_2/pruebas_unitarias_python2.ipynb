## 1. ¿Qué son las pruebas unitarias? <a id='intro'></a>

Una **prueba unitaria** verifica que una unidad pequeña de código (generalmente una función o método) funciona correctamente de forma **aislada**.

### ¿Por qué escribir pruebas?

| Beneficio | Descripción |
|-----------|-------------|
| ✅ Confianza | Sabes que tu código funciona |
| 🔄 Refactoring seguro | Puedes cambiar código sin miedo |
| 📖 Documentación viva | Las pruebas describen cómo funciona el código |
| 🐛 Detección temprana | Encuentras errores antes de producción |

## 2. Módulo `unittest` - Conceptos básicos <a id='conceptos'></a>

Python incluye `unittest` en su librería estándar. No necesitas instalar nada extra.

### Estructura básica de una prueba

```python
import unittest

class MisPruebas(unittest.TestCase):   # Hereda de TestCase
    
    def setUp(self):                    # Se ejecuta ANTES de cada prueba
        pass
    
    def tearDown(self):                 # Se ejecuta DESPUÉS de cada prueba
        pass
    
    def test_algo(self):               # ⚠️ DEBE empezar con 'test_'
        resultado = 2 + 2
        self.assertEqual(resultado, 4)

if __name__ == '__main__':
    unittest.main()
```

### Métodos de aserción más usados

| Método | Descripción |
|--------|-------------|
| `assertEqual(a, b)` | Verifica que `a == b` |
| `assertNotEqual(a, b)` | Verifica que `a != b` |
| `assertTrue(x)` | Verifica que `x` es verdadero |
| `assertFalse(x)` | Verifica que `x` es falso |
| `assertIsNone(x)` | Verifica que `x` es `None` |
| `assertRaises(exc, func)` | Verifica que se lanza una excepción |
| `assertIn(a, b)` | Verifica que `a` está en `b` |
| `assertAlmostEqual(a, b)` | Verifica igualdad con tolerancia (útil para floats) |

## 3. Ejemplo: Calculadora <a id='calculadora'></a>

Vamos a crear una calculadora simple y luego probarla paso a paso.

````python
# ============================================================
#  MÓDULO: calculadora.py  (lo definimos aquí para Colab)
# ============================================================

class Calculadora:
    """Calculadora básica con operaciones aritméticas."""

    def sumar(self, a, b):
        """Retorna la suma de a y b."""
        return a + b

    def restar(self, a, b):
        """Retorna la diferencia de a y b."""
        return a - b

    def multiplicar(self, a, b):
        """Retorna el producto de a y b."""
        return a * b

    def dividir(self, a, b):
        """Retorna la división de a entre b."""
        if b == 0:
            raise ValueError("No se puede dividir entre cero")
        return a / b

    def potencia(self, base, exponente):
        """Retorna base elevada al exponente."""
        return base ** exponente

    def es_par(self, numero):
        """Retorna True si el número es par."""
        return numero % 2 == 0


print("✅ Clase Calculadora definida correctamente")
````

````python
import unittest

class TestCalculadora(unittest.TestCase):

    # --- setUp: se ejecuta antes de CADA test ---
    def setUp(self):
        self.calc = Calculadora()
        print(f"
▶ Ejecutando: {self._testMethodName}")

    # ==================== SUMA ====================
    def test_suma_enteros_positivos(self):
        resultado = self.calc.sumar(3, 4)
        self.assertEqual(resultado, 7)

    def test_suma_con_negativo(self):
        resultado = self.calc.sumar(-1, 5)
        self.assertEqual(resultado, 4)

    def test_suma_flotantes(self):
        resultado = self.calc.sumar(0.1, 0.2)
        # ⚠️ Con floats usamos assertAlmostEqual para evitar problemas de precisión
        self.assertAlmostEqual(resultado, 0.3, places=5)

    # ==================== RESTA ====================
    def test_resta_resultado_positivo(self):
        self.assertEqual(self.calc.restar(10, 3), 7)

    def test_resta_resultado_negativo(self):
        self.assertEqual(self.calc.restar(3, 10), -7)

    # ==================== MULTIPLICACIÓN ====================
    def test_multiplicar_positivos(self):
        self.assertEqual(self.calc.multiplicar(4, 5), 20)

    def test_multiplicar_por_cero(self):
        self.assertEqual(self.calc.multiplicar(100, 0), 0)

    def test_multiplicar_negativos(self):
        self.assertEqual(self.calc.multiplicar(-3, -4), 12)

    # ==================== DIVISIÓN ====================
    def test_dividir_resultado_exacto(self):
        self.assertEqual(self.calc.dividir(10, 2), 5)

    def test_dividir_resultado_flotante(self):
        self.assertAlmostEqual(self.calc.dividir(1, 3), 0.3333, places=3)

    def test_dividir_entre_cero_lanza_excepcion(self):
        # Verificamos que se lanza el error correcto
        with self.assertRaises(ValueError) as context:
            self.calc.dividir(10, 0)
        # También podemos verificar el mensaje de error
        self.assertIn("cero", str(context.exception))

    # ==================== POTENCIA ====================
    def test_potencia_positiva(self):
        self.assertEqual(self.calc.potencia(2, 10), 1024)

    def test_potencia_cero(self):
        self.assertEqual(self.calc.potencia(5, 0), 1)

    # ==================== ES PAR ====================
    def test_es_par_con_par(self):
        self.assertTrue(self.calc.es_par(4))

    def test_es_par_con_impar(self):
        self.assertFalse(self.calc.es_par(7))

    def test_es_par_con_cero(self):
        self.assertTrue(self.calc.es_par(0))


# --- Ejecutar en Colab / Jupyter ---
loader = unittest.TestLoader()
suite = loader.loadTestsFromTestCase(TestCalculadora)
runner = unittest.TextTestRunner(verbosity=2)
resultado = runner.run(suite)
````

## 4. Uso de `Mock` <a id='mock'></a>

### ¿Qué es un Mock?

Un **Mock** es un objeto falso que simula el comportamiento de una dependencia real (base de datos, API externa, sistema de archivos, etc.).

```
SIN mock:  función → [API externa] → resultado real
CON mock:  función → [Mock que finge ser la API] → resultado controlado
```

### ¿Cuándo usar Mock?

- Cuando la función llama a una **API externa**
- Cuando depende de una **base de datos**
- Cuando el resultado depende del **tiempo** (`datetime.now()`)
- Cuando quieres probar el comportamiento ante **errores de red**
- Cuando la dependencia es **lenta o costosa**


````python
# ============================================================
# EJEMPLO 1: Mock básico - Simular una API de tipo de cambio
# ============================================================
from unittest.mock import Mock, patch, MagicMock

# --- Código a probar ---
class ServicioMoneda:
    """Servicio que consulta tipo de cambio desde una API externa."""

    def __init__(self, api_client):
        self.api_client = api_client

    def convertir_usd_a_cop(self, dolares):
        """Convierte dólares a pesos colombianos."""
        tasa = self.api_client.obtener_tasa("USD", "COP")
        return dolares * tasa

    def precio_con_iva(self, dolares):
        """Convierte y aplica IVA del 19%."""
        pesos = self.convertir_usd_a_cop(dolares)
        return pesos * 1.19


# --- Pruebas con Mock ---
class TestServicioMoneda(unittest.TestCase):

    def setUp(self):
        # Creamos un mock que reemplaza al cliente de API real
        self.api_mock = Mock()
        self.servicio = ServicioMoneda(self.api_mock)

    def test_conversion_usd_a_cop(self):
        # Configuramos qué retorna el mock
        self.api_mock.obtener_tasa.return_value = 4200

        resultado = self.servicio.convertir_usd_a_cop(10)

        self.assertEqual(resultado, 42000)
        # Verificamos que el mock fue llamado correctamente
        self.api_mock.obtener_tasa.assert_called_once_with("USD", "COP")

    def test_precio_con_iva(self):
        self.api_mock.obtener_tasa.return_value = 4000

        resultado = self.servicio.precio_con_iva(1)

        # 1 USD × 4000 × 1.19 = 4760
        self.assertAlmostEqual(resultado, 4760.0, places=1)

    def test_conversion_cuando_api_falla(self):
        # Simulamos un error de red
        self.api_mock.obtener_tasa.side_effect = ConnectionError("Sin conexión")

        with self.assertRaises(ConnectionError):
            self.servicio.convertir_usd_a_cop(5)

    def test_mock_llamado_correcta_cantidad_de_veces(self):
        self.api_mock.obtener_tasa.return_value = 4200

        self.servicio.convertir_usd_a_cop(10)
        self.servicio.convertir_usd_a_cop(20)

        # Verificamos que se llamó exactamente 2 veces
        self.assertEqual(self.api_mock.obtener_tasa.call_count, 2)


suite = unittest.TestLoader().loadTestsFromTestCase(TestServicioMoneda)
unittest.TextTestRunner(verbosity=2).run(suite)
````

````python
# ============================================================
# EJEMPLO 2: patch() - Parchear una función en su módulo
# ============================================================
import datetime

# --- Código a probar ---
def obtener_saludo():
    """Retorna un saludo según la hora del día."""
    hora = datetime.datetime.now().hour
    if hora < 12:
        return "Buenos días"
    elif hora < 18:
        return "Buenas tardes"
    else:
        return "Buenas noches"


class TestSaludo(unittest.TestCase):

    @patch('datetime.datetime')  # Parcheamos datetime.datetime
    def test_saludo_en_la_manana(self, mock_datetime):
        mock_datetime.now.return_value = datetime.datetime(2024, 1, 1, 9, 0, 0)
        self.assertEqual(obtener_saludo(), "Buenos días")

    @patch('datetime.datetime')
    def test_saludo_en_la_tarde(self, mock_datetime):
        mock_datetime.now.return_value = datetime.datetime(2024, 1, 1, 15, 0, 0)
        self.assertEqual(obtener_saludo(), "Buenas tardes")

    @patch('datetime.datetime')
    def test_saludo_en_la_noche(self, mock_datetime):
        mock_datetime.now.return_value = datetime.datetime(2024, 1, 1, 21, 0, 0)
        self.assertEqual(obtener_saludo(), "Buenas noches")


suite = unittest.TestLoader().loadTestsFromTestCase(TestSaludo)
unittest.TextTestRunner(verbosity=2).run(suite)
````

````python
# ============================================================
# EJEMPLO 3: Mock con side_effect para retornar diferentes valores
# ============================================================

class ProcesadorPagos:
    """Procesa pagos usando una pasarela externa."""

    def __init__(self, pasarela):
        self.pasarela = pasarela

    def pagar(self, monto):
        respuesta = self.pasarela.cobrar(monto)
        if respuesta.get("exito"):
            return f"Pago aprobado: {respuesta['transaccion_id']}"
        return "Pago rechazado"

    def pagar_en_cuotas(self, monto, cuotas):
        cuota = monto / cuotas
        resultados = []
        for i in range(cuotas):
            resultados.append(self.pagar(cuota))
        return resultados


class TestProcesadorPagos(unittest.TestCase):

    def setUp(self):
        self.pasarela_mock = Mock()
        self.procesador = ProcesadorPagos(self.pasarela_mock)

    def test_pago_exitoso(self):
        self.pasarela_mock.cobrar.return_value = {
            "exito": True,
            "transaccion_id": "TXN-001"
        }
        resultado = self.procesador.pagar(50000)
        self.assertEqual(resultado, "Pago aprobado: TXN-001")

    def test_pago_rechazado(self):
        self.pasarela_mock.cobrar.return_value = {"exito": False}
        resultado = self.procesador.pagar(50000)
        self.assertEqual(resultado, "Pago rechazado")

    def test_pago_en_cuotas_con_respuestas_variadas(self):
        # side_effect con lista: cada llamada retorna el siguiente elemento
        self.pasarela_mock.cobrar.side_effect = [
            {"exito": True, "transaccion_id": "TXN-A"},
            {"exito": True, "transaccion_id": "TXN-B"},
            {"exito": False},
        ]
        resultados = self.procesador.pagar_en_cuotas(90000, 3)
        self.assertEqual(resultados[0], "Pago aprobado: TXN-A")
        self.assertEqual(resultados[1], "Pago aprobado: TXN-B")
        self.assertEqual(resultados[2], "Pago rechazado")


suite = unittest.TestLoader().loadTestsFromTestCase(TestProcesadorPagos)
unittest.TextTestRunner(verbosity=2).run(suite)
````

## Descripción

Vas a probar un sistema bancario que gestiona cuentas de usuario. El sistema depende de:
- Un **repositorio** (base de datos simulada)

````python
# ============================================================
# CÓDIGO DEL TALLER - Sistema bancario
# Lee y entiende bien este código antes de empezar
# ============================================================

from unittest.mock import Mock, patch, call
from datetime import datetime


class CuentaNoEncontradaError(Exception):
    pass

class SaldoInsuficienteError(Exception):
    pass

class TransaccionBloqueadaError(Exception):
    pass


class CuentaBancaria:
    """Entidad que representa una cuenta bancaria."""

    def __init__(self, numero, titular, saldo=0.0):
        self.numero = numero
        self.titular = titular
        self.saldo = saldo
        self.activa = True
        self.historial = []

    def registrar_movimiento(self, tipo, monto):
        self.historial.append({
            "tipo": tipo,
            "monto": monto,
            "fecha": datetime.now().isoformat()
        })


class ServicioBancario:
    """
    Servicio principal que orquesta operaciones bancarias.
    
    Dependencias:
      - repositorio: CRUD de cuentas (base de datos)
      - notificador: envío de alertas al cliente
      - antifraude: validación de transacciones sospechosas
    """

    LIMITE_SIN_VERIFICACION = 1_000_000  # COP

    def __init__(self, repositorio, notificador, antifraude):
        self.repo = repositorio
        self.notificador = notificador
        self.antifraude = antifraude

    def depositar(self, numero_cuenta, monto):
        """
        Deposita monto en la cuenta. Notifica al titular.
        Retorna el nuevo saldo.
        """
        if monto <= 0:
            raise ValueError("El monto debe ser positivo")

        cuenta = self.repo.buscar(numero_cuenta)
        if not cuenta:
            raise CuentaNoEncontradaError(f"Cuenta {numero_cuenta} no existe")
        if not cuenta.activa:
            raise ValueError("La cuenta está inactiva")

        cuenta.saldo += monto
        cuenta.registrar_movimiento("deposito", monto)
        self.repo.guardar(cuenta)
        self.notificador.enviar(
            cuenta.titular,
            f"Depósito de ${monto:,.0f} COP. Saldo: ${cuenta.saldo:,.0f} COP"
        )
        return cuenta.saldo

    def retirar(self, numero_cuenta, monto):
        """
        Retira monto de la cuenta.
        Si el monto supera LIMITE_SIN_VERIFICACION, consulta antifraude.
        Retorna el nuevo saldo.
        """
        if monto <= 0:
            raise ValueError("El monto debe ser positivo")

        cuenta = self.repo.buscar(numero_cuenta)
        if not cuenta:
            raise CuentaNoEncontradaError(f"Cuenta {numero_cuenta} no existe")
        if not cuenta.activa:
            raise ValueError("La cuenta está inactiva")
        if cuenta.saldo < monto:
            raise SaldoInsuficienteError(
                f"Saldo insuficiente. Disponible: ${cuenta.saldo:,.0f}"
            )

        # Verificación antifraude para montos grandes
        if monto > self.LIMITE_SIN_VERIFICACION:
            aprobado = self.antifraude.verificar(numero_cuenta, monto)
            if not aprobado:
                raise TransaccionBloqueadaError("Transacción bloqueada por antifraude")

        cuenta.saldo -= monto
        cuenta.registrar_movimiento("retiro", monto)
        self.repo.guardar(cuenta)
        self.notificador.enviar(
            cuenta.titular,
            f"Retiro de ${monto:,.0f} COP. Saldo: ${cuenta.saldo:,.0f} COP"
        )
        return cuenta.saldo

    def transferir(self, numero_origen, numero_destino, monto):
        """
        Transfiere monto entre cuentas.
        Ambas cuentas deben existir y estar activas.
        Retorna (saldo_origen, saldo_destino).
        """
        cuenta_origen = self.repo.buscar(numero_origen)
        cuenta_destino = self.repo.buscar(numero_destino)

        if not cuenta_origen:
            raise CuentaNoEncontradaError(f"Cuenta origen {numero_origen} no existe")
        if not cuenta_destino:
            raise CuentaNoEncontradaError(f"Cuenta destino {numero_destino} no existe")
        if cuenta_origen.saldo < monto:
            raise SaldoInsuficienteError("Saldo insuficiente en cuenta origen")

        # Verificación antifraude para montos grandes
        if monto > self.LIMITE_SIN_VERIFICACION:
            aprobado = self.antifraude.verificar(numero_origen, monto)
            if not aprobado:
                raise TransaccionBloqueadaError("Transferencia bloqueada por antifraude")

        cuenta_origen.saldo -= monto
        cuenta_destino.saldo += monto
        cuenta_origen.registrar_movimiento("transferencia_salida", monto)
        cuenta_destino.registrar_movimiento("transferencia_entrada", monto)
        self.repo.guardar(cuenta_origen)
        self.repo.guardar(cuenta_destino)

        self.notificador.enviar(
            cuenta_origen.titular,
            f"Transferencia enviada: ${monto:,.0f} a {cuenta_destino.titular}"
        )
        self.notificador.enviar(
            cuenta_destino.titular,
            f"Transferencia recibida: ${monto:,.0f} de {cuenta_origen.titular}"
        )
        return cuenta_origen.saldo, cuenta_destino.saldo

    def consultar_saldo(self, numero_cuenta):
        """Retorna el saldo actual de la cuenta."""
        cuenta = self.repo.buscar(numero_cuenta)
        if not cuenta:
            raise CuentaNoEncontradaError(f"Cuenta {numero_cuenta} no existe")
        return cuenta.saldo


print("✅ Código del taller cargado correctamente")
print("
📌 Clases disponibles:")
print("  - CuentaBancaria(numero, titular, saldo)")
print("  - ServicioBancario(repositorio, notificador, antifraude)")
print("
📌 Excepciones:")
print("  - CuentaNoEncontradaError")
print("  - SaldoInsuficienteError")
print("  - TransaccionBloqueadaError")
````

````python
# ✅ SOLUCIÓN PARTE 2

class TestServicioBancarioParte2SOLUCION(unittest.TestCase):

    def setUp(self):
        self.repo = Mock()
        self.notificador = Mock()
        self.antifraude = Mock()
        self.servicio = ServicioBancario(self.repo, self.notificador, self.antifraude)

        self.cuenta_a = CuentaBancaria("A001", "Laura Ruiz", saldo=5_000_000)
        self.cuenta_b = CuentaBancaria("B001", "Carlos Mesa", saldo=100_000)

    # --- NOTIFICACIONES ---
    def test_deposito_notifica_al_titular(self):
        self.repo.buscar.return_value = self.cuenta_a
        self.servicio.depositar("A001", 500_000)
        self.notificador.enviar.assert_called_once()
        args = self.notificador.enviar.call_args[0]
        self.assertEqual(args[0], "Laura Ruiz")
        self.assertIn("500.000", args[1])

    def test_retiro_notifica_con_nuevo_saldo(self):
        self.repo.buscar.return_value = self.cuenta_a
        self.servicio.retirar("A001", 1_000)
        args = self.notificador.enviar.call_args[0]
        self.assertIn("4.999.000", args[1])

    # --- ANTIFRAUDE ---
    def test_retiro_grande_consulta_antifraude(self):
        self.repo.buscar.return_value = self.cuenta_a
        self.antifraude.verificar.return_value = True
        self.servicio.retirar("A001", 2_000_000)
        self.antifraude.verificar.assert_called_once_with("A001", 2_000_000)

    def test_retiro_grande_bloqueado_por_antifraude(self):
        self.repo.buscar.return_value = self.cuenta_a
        self.antifraude.verificar.return_value = False
        with self.assertRaises(TransaccionBloqueadaError):
            self.servicio.retirar("A001", 2_000_000)

    def test_retiro_pequeno_no_consulta_antifraude(self):
        self.repo.buscar.return_value = self.cuenta_a
        self.servicio.retirar("A001", 500_000)  # menor al límite
        self.antifraude.verificar.assert_not_called()

    # --- TRANSFERENCIAS ---
    def test_transferencia_notifica_a_ambas_partes(self):
        self.repo.buscar.side_effect = [self.cuenta_a, self.cuenta_b]
        self.servicio.transferir("A001", "B001", 100_000)
        self.assertEqual(self.notificador.enviar.call_count, 2)
        llamadas = self.notificador.enviar.call_args_list
        titulares = [c[0][0] for c in llamadas]
        self.assertIn("Laura Ruiz", titulares)
        self.assertIn("Carlos Mesa", titulares)

    def test_transferencia_mueve_saldo_correctamente(self):
        self.repo.buscar.side_effect = [self.cuenta_a, self.cuenta_b]
        saldo_origen, saldo_destino = self.servicio.transferir("A001", "B001", 500_000)
        self.assertEqual(saldo_origen, 4_500_000)
        self.assertEqual(saldo_destino, 600_000)

    def test_transferencia_destino_inexistente(self):
        self.repo.buscar.side_effect = [self.cuenta_a, None]
        with self.assertRaises(CuentaNoEncontradaError):
            self.servicio.transferir("A001", "X999", 10_000)


suite = unittest.TestLoader().loadTestsFromTestCase(TestServicioBancarioParte2SOLUCION)
unittest.TextTestRunner(verbosity=2).run(suite)
````

````python
# ✅ SOLUCIÓN PARTE 3

class TestServicioBancarioParte3SOLUCION(unittest.TestCase):

    def setUp(self):
        self.repo = Mock()
        self.notificador = Mock()
        self.antifraude = Mock()
        self.servicio = ServicioBancario(self.repo, self.notificador, self.antifraude)
        self.cuenta = CuentaBancaria("C001", "Pedro León", saldo=5_000_000)
        self.repo.buscar.return_value = self.cuenta

    def test_retiro_exactamente_en_el_limite_no_verifica_antifraude(self):
        # El límite es 1.000.000 — exactamente igual NO debería activar antifraude
        # (según la condición: monto > LIMITE)
        self.servicio.retirar("C001", 1_000_000)
        self.antifraude.verificar.assert_not_called()

    def test_retiro_un_peso_sobre_limite_verifica_antifraude(self):
        self.antifraude.verificar.return_value = True
        self.servicio.retirar("C001", 1_000_001)
        self.antifraude.verificar.assert_called_once()

    def test_historial_deposito_y_retiro(self):
        self.servicio.depositar("C001", 200_000)
        self.servicio.retirar("C001", 100_000)
        self.assertEqual(len(self.cuenta.historial), 2)
        self.assertEqual(self.cuenta.historial[0]["tipo"], "deposito")
        self.assertEqual(self.cuenta.historial[1]["tipo"], "retiro")

    def test_repo_falla_al_guardar_propaga_error(self):
        self.repo.guardar.side_effect = IOError("Error de base de datos")
        with self.assertRaises(IOError):
            self.servicio.depositar("C001", 10_000)

    def test_multiples_depositos_acumulan_saldo(self):
        saldo_inicial = self.cuenta.saldo
        for i in range(5):
            self.servicio.depositar("C001", 100_000)
        self.assertEqual(self.cuenta.saldo, saldo_inicial + 500_000)

    def test_guardar_llamado_con_objeto_correcto(self):
        self.servicio.depositar("C001", 50_000)
        # Verificamos que se guardó exactamente la cuenta modificada
        self.repo.guardar.assert_called_with(self.cuenta)

    def test_deposito_registra_monto_correcto_en_historial(self):
        self.servicio.depositar("C001", 777_000)
        self.assertEqual(self.cuenta.historial[-1]["monto"], 777_000)


suite = unittest.TestLoader().loadTestsFromTestCase(TestServicioBancarioParte3SOLUCION)
unittest.TextTestRunner(verbosity=2).run(suite)
````
