## TEST 2 — Contraseña muy corta

````python
%%ipytest -v

def test_contrasena_muy_corta():
    """
    DADO:    una contraseña de menos de 8 caracteres
    CUANDO:  llamo a validar_contrasena
    ENTONCES: es_valida=False y hay un error sobre la longitud
    """
    resultado = validar_contrasena('Ab1!')

    assert resultado['es_valida'] == False
    assert len(resultado['errores']) >= 1
    # Verifica que el mensaje mencione la longitud mínima
    assert any('8' in error or 'longitud' in error.lower() or 'minimo' in error.lower()
               for error in resultado['errores'])
````
