# Taller – Smoke, Regression y Acceptance Testing
Notebook listo para ejecutar en Google Colab.


```python
class Biblioteca:
    def __init__(self):
        self.libros={'python':{'disponible':True},'algoritmos':{'disponible':True},'base_datos':{'disponible':True}}
        self.usuarios={'juan':{'bloqueado':False,'multa':0},'ana':{'bloqueado':False,'multa':0}}
        self.prestamos={}

    def consultar_disponibilidad(self,libro):
        return self.libros.get(libro,{}).get('disponible',False)

    def registrar_prestamo(self,usuario,libro):
        if self.usuarios[usuario]['bloqueado']:
            return False
        if not self.libros[libro]['disponible']:
            return False
        self.libros[libro]['disponible']=False
        self.prestamos[libro]=usuario
        return True

    def devolver_libro(self,libro):
        if libro in self.prestamos:
            self.libros[libro]['disponible']=True
            del self.prestamos[libro]
            return True
        return False

    def calcular_multa(self,dias):
        if dias<=0:
            return 0
        return dias*500

    def pagar_multa(self,usuario,valor):
        if valor>=self.usuarios[usuario]['multa']:
            self.usuarios[usuario]['multa']=0
            self.usuarios[usuario]['bloqueado']=False
            return True
        return False
```

```python
sistema=Biblioteca()
```

```python
def smoke_tests():
    print('SMOKE TESTS')
    assert sistema.consultar_disponibilidad('python')==True
    assert sistema.registrar_prestamo('juan','python')==True
    assert sistema.consultar_disponibilidad('python')==False
    assert sistema.devolver_libro('python')==True
    assert sistema.calcular_multa(3)==1500
    print('Smoke tests OK')
```

```python
def regression_tests():
    print('REGRESSION TESTS')
    assert sistema.calcular_multa(1)==500
    assert sistema.calcular_multa(5)==2500
    sistema.registrar_prestamo('juan','algoritmos')
    assert sistema.libros['algoritmos']['disponible']==False
    sistema.devolver_libro('algoritmos')
    assert sistema.libros['algoritmos']['disponible']==True
    assert sistema.calcular_multa(0)==0
    print('Regression tests OK')
```

```python
def acceptance_tests():
    print('ACCEPTANCE TESTS')
    sistema.registrar_prestamo('ana','base_datos')
    assert sistema.consultar_disponibilidad('base_datos')==False
    sistema.devolver_libro('base_datos')
    assert sistema.consultar_disponibilidad('base_datos')==True
    multa=sistema.calcular_multa(3)
    assert multa==1500
    sistema.usuarios['ana']['multa']=1500
    sistema.usuarios['ana']['bloqueado']=True
    sistema.pagar_multa('ana',1500)
    assert sistema.usuarios['ana']['bloqueado']==False
    assert sistema.consultar_disponibilidad('python')==True
    print('Acceptance tests OK')
```

```python
def pipeline():
    smoke_tests()
    regression_tests()
    acceptance_tests()
    print('TODAS LAS PRUEBAS PASARON')

pipeline()
```
